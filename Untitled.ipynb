```python
import sys
import os
from os.path import join
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import proplot as pplt

pplt.rc['grid'] = False
pplt.rc['cmap.discrete'] = False
pplt.rc['cmap.sequential'] = 'viridis'
```

```python
folder = 'data/_output/'
bunch_filename = 'run_MEBT123_HZ04_VT34a_bunch_200.txt'
hist_filename = 'run_MEBT123_HZ04_VT34a_hist_200.txt'
```

```python
def load_bunch(filename, dframe=False):
    df = pd.read_table(filename, sep=' ', skiprows=14, usecols=range(0, 6), 
                       names=['x', 'xp', 'y', 'yp', 'z', 'dE'])
    if dframe:
        return df
    return df.values

def read_history(filename):
    file = open(filename, 'r')
    names = file.readlines()[0].split(',')
    names = [name.lstrip().split('[')[0].rstrip() for name in names]
    file.close()
    df = pd.read_table(filename, sep=' ', skiprows=1, names=names, usecols=range(len(names)))
    return df
```

```python
X = load_bunch(join(folder, bunch_filename))
X = X * 1000.0  # mm, mrad, mm, mrad, mm, MeV
```

```python
sys.path.append('/Users/46h/Dropbox (ORNL)/projects/scbd/')
from scdist.tools.plotting import corner
```

```python
labels = ["x [mm]", "xp [mrad]", "y [mm]", "yp [mrad]", "z [mm]", "w [MeV]"]
axes = corner(X[:, :], grid_kws=dict(labels=labels), cmap=pplt.Colormap('mono', left=0.05), thresh=0)
```

```python
fig, axes = pplt.subplots(ncols=3, share=False, figwidth=6, space=6)
for ax, (i, j) in zip(axes, [(0, 1), (2, 3), (4, 5)]):
    sns.histplot(x=X[:, i], y=X[:, j], bins=100, ax=ax, cmap=pplt.Colormap('mono', left=0.05),
                 ec='None')
    ax.format(xlabel=labels[i], ylabel=labels[j])
plt.savefig('figures/2dproj.png')
```

```python
history = read_history(join(folder, hist_filename))
history
```

```python
fig, ax = pplt.subplots(figsize=(5, 2))
ax.plot(history['s'], history[['xrms', 'yrms']])
ax.format(ylabel='[mm]', xlabel='s [mm]')
ax.legend(ncols=1)
plt.savefig('figures/rms.png')
plt.show()
```
